```python
import pandas as pd
dataset=pd.read_csv("train.csv")
```

```python
dataset.head()
```

```python
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.isna().sum()
```

```python
dataset["Order Date"]=pd.to_datetime(dataset["Order Date"],errors='coerce')
```

```python
dataset.head()
```

```python
dataset["Ship Date"]=pd.to_datetime(dataset["Ship Date"],errors='coerce')
```

```python
dataset.head()
```

```python
dataset["Month"]=dataset["Order Date"].dt.month
```

```python
dataset.head()
```

```python
dataset["Year"]=dataset["Order Date"].dt.year
```

```python
dataset.head()
```

```python
monthly_sales=dataset.groupby(["Year","Month"])["Sales"].sum().reset_index()
```

```python
monthly_sales.max()
```

```python
monthly_sales.min()
```

```python
monthly_sales_sorted=monthly_sales.sort_values(["Year","Month"],ascending=[True,True])
```

```python
monthly_sales_sorted.head(5)
```

```python
monthly_sales_sorted.tail(5)
```

```python
monthly_sales_sorted
```

```python
yearly_sales=dataset.groupby("Year")["Sales"].sum().reset_index()
```

```python
yearly_sales
```

```python
monthly_sales_sorted["Change"]=monthly_sales_sorted["Sales"].pct_change()*100
```

```python
monthly_sales_sorted
```

```python
yearly_sales["Change"]=yearly_sales["Sales"].pct_change()*100
```

```python
yearly_sales
```

```python
categorical_sales=dataset.groupby("Category")["Sales"].sum().reset_index()
```

```python
categorical_sales["Share"]=categorical_sales["Sales"]/categorical_sales["Sales"].sum()*100
```

```python
categorical_sales
```

```python
subcategorical_sales=dataset.groupby("Sub-Category")["Sales"].sum().reset_index()
```

```python
subcategorical_sales_sorted=subcategorical_sales.sort_values(by="Sales")
```

```python
subcategorical_sales_sorted["Share"]=subcategorical_sales_sorted["Sales"]/subcategorical_sales_sorted["Sales"].sum()*100
```

```python
subcategorical_sales_sorted
```

```python
regional_sales=dataset.groupby("Region")["Sales"].sum().reset_index()
```

```python
regional_sales["Dominance"]=regional_sales["Sales"]/regional_sales["Sales"].sum()*100
```

```python
regional_sales
```
